```python
import pandas as pd
import os
import hvplot.pandas
from pathlib import Path
```

```python
weather_2015_df = pd.read_csv(Path("../desktop/weather_2015.csv"))
weather_2018_df = pd.read_csv(Path("../desktop/weather_2018.csv"))
weather_2016_df = pd.read_csv(Path("../desktop/weather_2016.csv"))
weather_2017_df = pd.read_csv(Path("../desktop/weather_2017.csv"))
weather_2019_df = pd.read_csv(Path("../desktop/weather_2019.csv"))
```

```python
weather_2015_df = weather_2015_df.groupby('id_station').mean('temp_avg')
```

```python
# Review the resulting DataFrame
display(weather_2015_df.head())
display(weather_2015_df.tail())
```

```python
weather_df = pd.concat([weather_2015_df, weather_2016_df, weather_2017_df, weather_2018_df, weather_2019_df], axis=0, sort=False)
```

```python
# Review the resulting DataFrame
display(weather_df.head())
display(weather_df.tail())
```

```python
weather_df = weather_df.groupby('date').mean('temp_avg')
```
