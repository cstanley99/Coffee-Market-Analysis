```python
import os
import pandas as pd
from pathlib import Path
import hvplot.pandas
import requests
import yfinance as yf
%matplotlib inline
```

# Coffee Prices 

The ICO composite indicator price, a historical series which can be extended back to 1947,
provides an overall benchmark for the price of green coffee of all major origins and type, 
considered to be the best available measure of levels of green coffee transactions on a global basis.
(soure: www.ico.org)

```python
ico_coffee_price = pd.read_csv(
    Path("ICO Prices 1990-2019.csv"),
    names=['Months','ICO Composite','Colombian Milds','Other Milds','Brazilian Naturals','Robustas'])
```

```python
ico_coffee_price.dropna(inplace=True)
display(ico_coffee_price.head(5))
display(ico_coffee_price.tail(5))
```

```python
ico_coffee_price.set_index('Months')
```

```python
ico_coffee_price.set_index(pd.date_range(start='1990-01', end='2020-01',freq='M'),inplace=True)
```

```python
display(ico_coffee_price.head(5))
display(ico_coffee_price.tail(5))
```

```python
ico_coffee_price.drop(columns=["Months"],inplace=True)
```

```python
display(ico_coffee_price.head(5))
display(ico_coffee_price.tail(5))
```

```python
ico_coffee_price = ico_coffee_price.loc["2015-01-31":"2019-12-31"]
```

```python
display(ico_coffee_price.head(5))
display(ico_coffee_price.tail(5))
```

```python
ico_coffee_price.hvplot.line(
    value_label='US Cents per Pound',
    height=400, 
    width=700, 
    legend='top', 
    title='International Coffee Organization Pricing'
)
```

```python
futures_data = yf.download("KC=F", start="2015-01-01", end="2019-12-01",interval = "1mo")
```

```python
futures_data.head(5)
```

```python
futures_data.drop(columns=["Open","High","Low","Adj Close","Volume"],inplace=True)
```

```python
futures_data.head(5)
```

```python
futures_data.hvplot.line(
    value_label='US Cents per Pound',
    height=400, 
    width=700, 
    legend='top', 
    title='Futures Pricing'
)
```

```python
ico_only = ico_coffee_price.hvplot.line(
    y='ICO Composite',
    value_label='US Cents per Pound',
    height=400, 
    width=700, 
    legend='top', 
    title='ICO and Futures Prices'
)
```
